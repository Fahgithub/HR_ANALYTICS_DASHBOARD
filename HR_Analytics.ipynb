```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'data' is your DataFrame

# Group data by 'Education Field' and count occurrences
education_counts = data.groupby('Education Field')['Employee Number'].count()

# Extract the education fields and counts
education_fields = education_counts.index.tolist()
employee_counts = education_counts.values.tolist()

# Create the bar chart
# Removed the 'height=6' argument as 'employee_counts' already provides the heights
plt.bar(education_fields, employee_counts)
plt.xlabel("Education Field")
plt.ylabel("Employee Number")
plt.title("Employee Count by Education Field")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.show()
```

```python
plt.figure(figsize=(8, 5))
Gender = data.groupby('Gender')['Employee Number'].count()
education_fields =Gender.index.tolist()
employee_counts =Gender.values.tolist()
plt.bar(education_fields,employee_counts, color=['pink', 'cyan'])
plt.title("Number of Employees as per Gender")
plt.xlabel("Gender")
plt.ylabel("Employee")
plt.ylim(0,1000)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'data' is your DataFrame

# Group data by 'Education Field' and count occurrences
education_counts = data.groupby('Business Travel')['Employee Number'].count()

# Extract the education fields and counts
education_fields = education_counts.index.tolist()
employee_counts = education_counts.values.tolist()

# Create the bar chart
# Removed the 'height=6' argument as 'employee_counts' already provides the heights
plt.bar(education_fields, employee_counts,color=['brown'])
plt.xlabel("Business Travel")
plt.ylabel("Employee Number")
plt.title("Employee Count by Business Travel")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'data' is your DataFrame

# Group data by 'Education Field' and count occurrences
education_counts = data.groupby('Department')['Employee Number'].count()

# Extract the education fields and counts
education_fields = education_counts.index.tolist()
employee_counts = education_counts.values.tolist()

# Create the bar chart
# Removed the 'height=6' argument as 'employee_counts' already provides the heights
plt.pie(employee_counts, labels=education_fields,autopct='%1.1f%%')
plt.title("Departments")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.show()
```
